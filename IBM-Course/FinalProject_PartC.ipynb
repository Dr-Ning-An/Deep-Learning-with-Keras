Download and clean dataset

```python
# import pandas and numpy libraries
import pandas as pd
import numpy as np
```

```python
# read data to pandas frame
concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data.shape
```

```python
# Split data into predictors and target
concrete_data_columns = concrete_data.columns
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
# Normalize the predictors
predictors_norm = (predictors - predictors.mean()) / predictors.std()
```

```python
# dimension of inputs
n_cols = predictors_norm.shape[1] # number of predictors
```

Build a Neural Network

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# build the model
model = regression_model()
```

```python
# train the model and evaluate the model
MSE = np.zeros(50) # Initialize an array for storing the mean-squared-error for each run
for i in range(50):
    from sklearn.model_selection import train_test_split
    # Randomly split the data into a training and test sets by holding 30% of the data for testing.
    predictors_norm_train, predictors_norm_test, target_train, target_test = train_test_split(predictors_norm, target, test_size=0.3, random_state=42)
    # fit the model
    model.fit(predictors_norm_train, target_train, epochs=100)
    # Evaluate the model on the test data and compute the mean squared error between the predictions and true data
    target_pred = model.predict(predictors_norm_test)
    from sklearn.metrics import mean_squared_error
    MSE[i] = mean_squared_error(target_pred, target_test)
    i = i + 1
```

#### Calculate the mean and the standard deviation of the mean squared errors

```python
np.mean(MSE) # mean of the mean-squared-errors
```

```python
np.std(MSE) # standard deviation of the mean-squared-errors
```
